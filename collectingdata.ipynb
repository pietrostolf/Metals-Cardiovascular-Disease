```python
# Dependencies and Setup
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib as matplot

# File to Load 
demographic_to_load = "Resources/demographic.csv"
diet_to_load = "Resources/diet.csv"
examination_to_load = "Resources/examination.csv"
labs_to_load = "Resources/labs.csv"
# Med_to_load = "Resources/medications.csv"
questto_load = "Resources/questionnaire.csv"


# # Read Files and store into Pandas DataFrames 
demo_data = pd.read_csv(demographic_to_load, encoding='utf-8')
diet_data = pd.read_csv(diet_to_load, encoding='utf-8')
exam_data = pd.read_csv(examination_to_load, encoding='utf-8')
lab_data = pd.read_csv(labs_to_load, encoding='utf-8')
# Med_data = pd.read_csv(Med_to_load) 
quest_data = pd.read_csv(questto_load, encoding='utf-8')

# read the byte string as a binary file
# with open('Med_to_load', 'rb') as f:
#     byte_string = f.read()

# # decode the byte string using utf-8
# try:
#     decoded_string = byte_string.decode('utf-8')
# except UnicodeDecodeError as e:
#     print(f"Error: {e}")
#     decoded_string = byte_string.decode('latin-1', errors='ignore')

# create a DataFrame from the decoded string
# df = pd.read_csv(pd.compat.StringIO(decoded_string), encoding='utf-8')


# Combine the data into a single dataset.  
first = pd.merge(demo_data, diet_data, how="left", on=["SEQN"])
second=pd.merge(first, exam_data, how="left", on=["SEQN"])
third=pd.merge(second, lab_data, how="left", on=["SEQN"])
complete_data=pd.merge(third, quest_data, how="left", on=["SEQN"])
complete_data.head()
```
