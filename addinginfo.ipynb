```python
# Dependencies and Setup
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib as matplot

# File to Load 
examination_to_load = "Resources/examination.csv"
labs_to_load = "Resources/labs.csv"


# Read Files and store into Pandas DataFrames 
exam_data = pd.read_csv(examination_to_load, encoding='utf-8')
lab_data = pd.read_csv(labs_to_load, encoding='utf-8')





# Combine the data into a single dataset.  

complete_data=pd.merge(lab_data, exam_data, how="left", on=["SEQN"])
complete_data.head()
```

```python
cols_to_clean = ['SEQN','LBXTHG', 'LBXBSE', 'LBXBPB', 'LBXBMN', 'LBXBCD', 'LBXTR', 'LBDLDL', 'LBDHDD','BPXDI1','BPXDI2','BPXSY1','BPXSY2','LBXGH']
variable_df=complete_data[cols_to_clean]
variable_df.head()
```

```python

df_cleaned = complete_data.dropna(axis=1, how='all')
df_cleaned= complete_data.dropna(axis=0, how='all')
df_cleaned.head()
```

```python
#dropping the rows with null cells
for col in cols_to_clean:
    variable_df = variable_df.dropna(subset=[col])
```

```python
 #renaming the variables
df2=variable_df.rename(columns={
    "SEQN":"patient ID",
    "LBXTHG":"mercury (ug/L)", 
    "LBXBSE":"selenium(ug/L)", 
    "LBXBPB":"lead(ug/dL)", 
    "LBXBMN": "manganese(ug/L)", 
    "LBXBCD":"cadmium(ug/L)", 
    "LBXTR":"triglycerides(mg/dL)", 
    "LBDLDL":"LDL-cholesterol(mg/dL)", 
    "LBDHDD":"HDL-Cholesterol(mg/dL)",
    "BPXDI1":"Diastolic Blood pressure (first reading) mmHg",
    "BPXDI2":"Diastolic Blood pressure (second reading) mmHg",
    "BPXSY1":"Systolic Blood pressure (first reading) mmHg",
    "BPXSY2":"Systolic Blood pressure (second reading) mmHg",
    "LBXGH":"Glycohemoglobin(%)",
    
})
```
